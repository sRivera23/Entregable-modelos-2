#Analisis de datos atípicos

```python
plt.figure(figsize=(12,9))

sns.boxplot(data=dfBody)
plt.title('Gráfico de Cajas')
plt.show()
```

```python
Q1 = dfBody.quantile(0.25)
Q3 = dfBody.quantile(0.75)
IQR = Q3 - Q1

# Filtrar valores atípicos
df_filtrado = dfBody[~((dfBody < (Q1 - 1.5 * IQR)) | (dfBody > (Q3 + 1.5 * IQR))).any(axis=1)]

targetBody = df_filtrado['class']
df_filtrado = df_filtrado.drop(columns='class', inplace=False)
df_filtrado.to_csv("df_bodyfat.csv", index=False)
targetBody.to_csv("target.csv", index=False)
```
